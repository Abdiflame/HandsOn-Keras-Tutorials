```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(inputs):
    sigmoid_scores = [ 1 / float(1 + np.exp(-x)) for x in inputs]
    return sigmoid_scores

sigmoid_inputs = [2, 3, 4, 5, 6]
print("Sigmoid Function Output: ", sigmoid(sigmoid_inputs))
```

```python
def line_graph(x ,y, x_title, y_title):
    plt.plot(x, y)
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.show()
    
graph_x = range(0,20)
graph_y = sigmoid(graph_x)

print("Graph X readings: {", (graph_x),  "} ")
print("Graph Y readings: {", (graph_y),  "} ")

line_graph(graph_x, graph_y, "Inputs", "Sigmoid Scores")
```
